```python
lending = lending.replace('-', ny.nan)
```

```python
lending['TA Ratio1'] = pd.to_numeric(lending['TA Ratio1'])
```

```python
lending
```

```python
lending['TA Ratio1'] = lending['TA Ratio1'].fillna(lending['TA Ratio1'].mean())
```

```python
lending
```

```python
lending['TBL Ratio1'] = lending['TBL Ratio1'].fillna(lending['TBL Ratio1'].mean())
```

```python
lending['Amount ($1,000)'] = lending['Amount ($1,000)'].fillna(lending['Amount ($1,000)'].mean())
```

```python
lending['Number '] = lending['Number '].fillna(lending['Number '].mean())
```

```python
lending
```

```python
lending["Lender Asset Size"] = lending["Lender Asset Size"].str.strip()
```

```python
size_map = {'>$10B': 0, '$10B-$50B': 1, '>$50B': 2}
```

```python
lending['Lender Asset Size'] = lending['Lender Asset Size'].map(size_map)
```

```python
lending = lending.replace('.', ny.nan)
```
