```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cafe_sales = pd.read_csv("dirty_cafe_sales.csv")
```

```python
cafe_sales
```

```python
cafe_sales.info()
```

```python
cafe_sales.isnull().sum()
```

```python
cafe_sales = cafe_sales.replace(["ERROR", "UNKNOWN"], py.nan)
```

```python
num_colums = ["Quantity", "Price Per Unit", "Total Spent"]

for x in num_colums:
    cafe_sales[x] = pd.to_numeric(cafe_sales[x])
```

```python
for x in num_colums:
    cafe_sales[x] = cafe_sales[x].fillna(cafe_sales[x].mean())
```

```python
cat_colums = ["Item", "Payment Method", "Location"]

for x in cat_colums:
    cafe_sales[x] = cafe_sales[x].fillna(cafe_sales[x].mode()[0])
```

```python
cafe_sales["Transaction Date"] = cafe_sales["Transaction Date"].fillna(cafe_sales["Transaction Date"].mode()[0])
```
